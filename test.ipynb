```python
# Define two vectors
v1 = np.array([3, 4])
v2 = np.array([1, 2])

# Dot product
dot_product = np.dot(v1, v2)

print(f"Dot Product of {v1} and {v2} is: {dot_product}")
```

```python
x = np.linspace(-10, 10, 100)
y = x**2  # Quadratic function

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.title("Quadratic Function Plot")
plt.grid()
plt.show()
```
